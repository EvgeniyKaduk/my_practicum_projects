**Описание проекта**

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Наша задача — обучить на этих данных модель для прогноза оттока клиентов. Необходимо получить метрику качества модели ROC-AUC не ниже 0.85 на тестовой выборке.


**Описание данных**

*Файлы с данными:*

`contract_new.csv` — информация о договоре;

`personal_new.csv` — персональные данные клиента;

`internet_new.csv` — информация об интернет-услугах;

`phone_new.csv` — информация об услугах телефонии.

*Все признаки в файлах:*

`customerID` — идентификатор абонента;

`BeginDate` — дата начала действия договора;

`EndDate` — дата окончания действия договора;

`Type` — тип оплаты: раз в год-два или ежемесячно;

`PaperlessBilling` — электронный расчётный лист;

`PaymentMethod` — тип платежа;

`MonthlyCharges` — расходы за месяц;

`TotalCharges` — общие расходы абонента;

`gender` — пол;

`SeniorCitizen` — является ли абонент пенсионером;

`Partner` — есть ли у абонента супруг или супруга;

`Dependents` — есть ли у абонента дети;

`InternetService` — тип подключения;

`OnlineSecurity` — блокировка опасных сайтов;

`OnlineBackup` — облачное хранилище файлов для резервного копирования данных;

`DeviceProtection` — антивирус;

`TechSupport` — выделенная линия технической поддержки;

`StreamingTV` — стриминговое телевидение;

`StreamingMovies` — каталог фильмов;

`MultipleLines` — подключение телефона к нескольким линиям одновременно.

Информация о договорах актуальна на 1 февраля 2020 года.

## Предобработка и исследовательский анализ данных

```python
# приведем названия столбцов к змеиному регистру
full_data = clean_columns(full_data)

# выведем первые строки датафрейма
full_data.head()
```

```python
# выведем строки со значениями в 'total_charges', которые нельзя перевести в цифровой формат
full_data[pd.to_numeric(full_data['total_charges'], errors='coerce').isna()]
```

```python
# удаляем строки с пробелами в столбце 'total_charges'
full_data = full_data.query('total_charges!=" "')
```

```python
# изменим тип данных в столбце 'total_charges'
full_data['total_charges'] = full_data['total_charges'].astype('float')
```

```python
# проверка на наличие дупликатов
full_data['customer_id'].duplicated().sum()
```

```python
# заполним пропуски по признаку 'internet_service'
full_data['internet_service'] = full_data['internet_service'].fillna('No internet')
```

```python
# заполним пропуски по признакам, связанным с интернет-услугами
full_data.iloc[:, 12:19] = full_data.iloc[:, 12:19].fillna('No')
```

```python
# заполним пропуски по признаку 'multiple_lines'
full_data['multiple_lines'] = full_data['multiple_lines'].fillna('No phone')
```

```python
# выводим первые строки датафрейма после заполнения пропусков
full_data.head(10)
```

```python
# посмотрим конечные даты договора с абонентами за исключением действующих договоров
full_data.query('end_date!="No"')['end_date'].sort_values(ascending=False).head()
```

```python
# заполним конечную дату договора с абонентом
full_data['end_date'] = full_data['end_date'].replace('No', '2020-02-01')
```

```python
# переведем данные по датам в формат  datetime
full_data['begin_date'] = pd.to_datetime(full_data['begin_date'], format='%Y-%m-%d')
full_data['end_date'] = pd.to_datetime(full_data['end_date'], format='%Y-%m-%d')
```

```python
# создадим новый признак - длительность действия договора с абонентом в днях
full_data['duration_in_days'] = (full_data['end_date'] - full_data['begin_date']).dt.days
```

```python
# добавляем  целевой признак в датафрейм
full_data['user_churn'] = (~full_data['end_date'].isin([pd.to_datetime('2020-02-01')])).astype('int')
```

```python
# удаляем из датафрейма строки, которые не будем использовать в прогнозной модели
full_data = full_data.drop(['customer_id', 'begin_date', 'end_date'], axis=1)
```

```python
 # создадим функцию для построения графиков для количественных признаков
def num_feature_plots(data, feature, bins, name):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
    fig.suptitle(name, fontsize=15)
    
    # строим диаграмму размаха
    sns.set_style("whitegrid")
    sns.boxplot(ax=axes[0], data = data, y = feature)
    axes[0].set_title(f'Диаграмма размаха для признака: {feature}')
    axes[0].set_ylabel(f'{feature}')
    
    # строим гистрограмму
    sns.histplot(ax=axes[1], data=data[feature], bins=bins)
    axes[1].set_title(f'Распределение значений признака: {feature}')
    axes[1].set_xlabel(f'{feature}')
    axes[1].set_ylabel('Частоты')
    plt.show()
```

```python
# определяем входные и целевой признаки для модели
X = full_data.drop(['user_churn'], axis=1)
y = full_data['user_churn']
# делим данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE,
                                                    random_state=RANDOM_STATE, stratify=y)
```

```python
# выделяем количественные и категориальные признаки
num_columns = ['duration_in_days', 'total_charges']
cat_columns = X_train.select_dtypes(include='object').columns
display(cat_columns)
```

```python
# задаем преобразование данных в ColumnTransformer
data_preprocessor = ColumnTransformer(
    [
        ('num', StandardScaler(), num_columns),
        ('ohe', OneHotEncoder(drop='first', sparse_output=True), cat_columns)
    ],
         remainder='passthrough',
         force_int_remainder_cols=False
)
```

**Вывод:**

На этапе подготовки данных мы выделили целевой и входные признаки для модели и провели разбивку данных на тренировочную и тестовую выборки. Т.к. у нас дисбаланс классов мы задаем стратификацию по целевому признаку в методе train_test_split. Затем мы выделяем количественные и категориальные признаки для их дальнейшего преобразования и подготовки к моделированию в ColumnTransformer. Признаки `duration_in_days`, `total_charges` - количественные. Признаки `payment_method`, `tech_support`, `streaming_movies`, `internet_service`, `multiple_lines`, `partner`, `online_backup`, `online_security`, `device_protection`, `paperless_billing`, `type`, `streaming_tv`, `senior_citizen` - категориальные. Количенственные признаки мы масштабируем с помощью StandardScaler, а категориальные кодируем с помощью OneHotEncoder в ColumnTransform. При этом признак `senior_citizen` мы не задаем как категораильный для обработки в OneHotEncoder, поскольку он уже по сути представлен в кодированном виде. После обработки в ColumnTransformer данные подготовлены для дальнейшего использования в пайплайне и подбора оптимальных гиперпараметров для модели.

## Обучение моделей машинного обучения 

```python
# создаем пайплайн для подбора оптимальной модели
best_model_pipe = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LogisticRegression(random_state=RANDOM_STATE))
    ]
)

# задаем словари гиперпараметров для двух моделей
param_grid = [
               # для модели LogisticRegression()
               {
                   'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear')],
                   'models__penalty': ['l1', 'l2'],
                   'models__C': [0.5, 3, 5, 7]
               },
               # для модели RidgeClassifier()
               {
                   'models': [RidgeClassifier()],
                   'models__alpha': [0.01, 0.1, 0.3, 0.5]
               },
               # для модели SVC()  
               {
                   'models': [SVC(random_state=RANDOM_STATE, kernel='rbf', probability=True)],
                   'models__C': [0.01, 0.1, 0.5, 1],
                   'models__gamma': [0.001, 0.01]
               }
             ]
```

```python
# инициализируем класс для автоматизированного поиска лучших гиперпараметров
# на данных кросс-валидационной выборки c помощью GridSearchCV
best_model_gs = GridSearchCV(best_model_pipe, param_grid=param_grid, cv=5, scoring='roc_auc',
                             n_jobs=-1, verbose=3)
best_model_gs.fit(X_train, y_train)
```

```python
# выводим параметры лучшей модели
print('Лучшая обученная модель и её параметры:\n\n', best_model_gs.best_params_)
```

```python
# выведем  общие результаты поиска гиперпараметров лучшей модели
print('\033[1mОбщие результаты поиска гиперпараметров лучшей модели:\033[0m')
display(pd.DataFrame(best_model_gs.cv_results_))
```

```python
# инициализируем модель LGBMClassifier
model = LGBMClassifier(objective="regression", random_state=RANDOM_STATE)

# задаём словарь гиперпараметров
param_gs = {
        'learning_rate': [0.2, 0.4],
        'max_depth': [10, 20, 25],
        'n_estimators': [50, 100]
            }

gs_lgbm = GridSearchCV(model, param_grid=param_gs, cv=5, scoring='roc_auc',
                        n_jobs=-1, verbose=3)
```

Графики и матрица корреляции показала, что признаки `gender`, `dependents`, `senior_citizen` очень слабо связаны целевым признаком, поэтому мы решили исключить эти признаки для будущего использования в модели классификации. Оставшиеся категориальные признаки показали наличие разной степени зависимости с целевым признаком. Например, мы видим явную зависимость между значениями `multiple_lines` и `user_churn`: доля клиентов, отказавшихся от услуг оператора, заметно выше среди клиентов, использующих подключение телефона к нескольким линиям одновременно, нежели среди клиентов, не использующих такого подключения. Также наблюдается заметная связь признаков `partner`, `payment_method`, `type`, `online_backup`, `device_protection` с целевым признаком. На основе корреляционного анализа мы исключили из датафрейма признак `monthly_charges`, поскольку прослеживается явная мультиколлениарность с признаком `internet_service` (коэффициент корреляции Фи 0.92) и кроме того `monthly_charges` имеет довольно сильную корреляцию с признаками `streaming_tv` (0.84) и `streaming_movies` (0.83).

Мы также изучили матрицы ошибок для тренировочной и тестовой выборок и заметили, что при заданном по умолчанию пороге классификации 0.5 выбранная оптимальная модель заметно лучше справляется с классификацией объектов класса 0 по сравнению с классификацией объектов класса 1, иными словами мы получили высокое значение метрики `precision` на тестовой выборке и гораздо худший результат метрики `recall`. Вероятно, для «ТелеДом» важнее будет полнее классифицировать клиентов класса 1, нежели с меньшей ошибкой присвоить класс 1. Для идеальной модели нужно подобрать какое-то компромиссное пороговое значение для классификации, так чтобы было хорошее значение охвата клиентов класса 1, но при этом не слишком низкое значение точности присвоения класса 1. В качестве лучшего значения порога мы выбрали 0.27, при котором на тестовой выборке получили значение метрики recall=0.71.

Полученная в результате исследования модель дает очень хорошее качество по метрике ROC-AUC. Данная метрика не зависит от порога классфикации, однако для лучшей классификации абонентов следует хорошо подумать о выборе оптимального порога классфикации: мы предложили один из вариантов, но при уточнении издержек результат может быть скорректирован. Компании «ТелеДом» нужно сопоставить издержки от потери каждого абонента с издержками от работы с абонентами (подготовка промокодов и спецпредложений), которые неправильно были классифицированы моделью как готовые разорвать договор, тогда как на самом деле таковыми не являются. В результате определения точных издержек наша модель может быть доработана, и мы можем правильным образом скорректировать значение порога для классификации.
